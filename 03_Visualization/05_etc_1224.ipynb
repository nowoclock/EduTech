# 기타 시각화

## Heatmap

- vmin(), vmax()로 최대 최소 범위를 설정 가능.
---
### 주요특징

- **색상으로 값의 크기 표현**
    - 값이 클수록 진한 색상, 작을수록 연한 색상으로 데이터를 표현한다.
- **행렬 시각화**
    - 2차원 데이터를 행렬 형태로 표현해서 데이터 간의 상관 관계를 시각적으로 표현한다.
- **적합한 데이터**
    - 수치형 데이터
---
### 한계

- **색상 해석**
    - 각 데이터를 격자(셀)에 해당하는 색상으로 표현하기 때문에 정확한 수치를 파악하기 어렵다.
- **대규모 데이터 해석**
    - 데이터가 많을 경우 행과 열, 격자가 많아지고, 해석이 어려워진다.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_df = sns.load_dataset('titanic')
titanic_df.info()
```

```python
# 상관관계 행렬 계산
# 수치형 데이터만 선택

numeric_df = titanic_df.select_dtypes(include=['int64','float64']).corr()

# 상관걔수 행렬 계산
corr_numeric = numeric_df.corr()

sns.heatmap(
    corr_numeric,
    annot = True, # 수치와 그래프 연결
    cmap = 'Greens'
)
```

## Pie Chart
---

### 주요특징

- **데이터의 비율 표현**
    - 데이터의 각 범주가 전체에서 차지하는 비율을 표현한다.
- **범주형 데이터 표현**
    - 범주형 데이터간 비율을 시각화하는데 적합하다.
- **간결하고 직관적인 표현**
    - 범주형 데이터가 각각 얼마나 차지하는지 직관적으로 파악할 수 있다.

---
### 한계

- **데이터 수 제한**
    - 부분 데이터가 많을수록 조각의 수도 늘어나기 때문에 모든 데이터를 표현하기 어렵다.

- **정확한 수치 비교 어려움**
    - 비율에 대한 레이블이 없으면 조각의 크기에 의존하여 각 데이터를 비교해야한다. 이 때 비슷한 비율의 가진 데이터 간에는 비교가 어렵다.

```python
import pandas as pd
```

```python
 df = pd.read_csv("/content/drive/MyDrive/PM_교육/DataSets/movie_ratings.csv")
```

```python
# 범주형 데이터의 비율 (빈도 수)
# value_counts()
values = df['Genre'].value_counts()

labels = values.index

plt.pie(
    values,
    labels = labels,
    autopct = '%1.1f%%'
)
plt.show()
```
