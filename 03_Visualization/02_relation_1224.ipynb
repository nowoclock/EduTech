```python
import seaborn as sns
import matplotlib.pyplot as plt

# 샘플 데이터 로드
tips = sns.load_dataset("tips")

# 팁 금액과 총 청구금액의 연관성 시각화
sns.scatterplot(data=tips, x="total_bill", y="tip")

# 그래프 제목 추가
plt.title("Tip Amount And Total Bill")

# 그래프 보여주기
plt.show()
```

## Scatter Plot 산점도

### **주요 특징**

- **데이터의 분포와 관계 파악**
    - 두 변수 간의 관계를 시각적으로 파악할 수 있다.
    - 예를 들어 변수 간에 강한 상관 관계가 있는지 아니면 관계가 없는지를 확인할 수 있다.
- **이상치 확인**
    - 데이터의 이상치나 특이한 패턴을 시각적으로 식별하기 쉽다.
- **데이터 밀도**
    - 많은 수의 데이터 포인트가 특정 영역에 모여 있으면 해당 영역에서 데이터의 밀도가 높다는 것을 의미한다.
- **색상 및 크기**
    - 산점도에서 점의 색상이나 크기를 조절하여 추가 변수의 정보를 표현할 수 있다.
- **적합한 데이터**
    - 두 개의 연속형 데이터
    - 두 개의 연속형 데이터를 구분하는 범주형 데이터
---
### **한계**

- **대량의 데이터**
    - 산점도의 주요 위험은 과도한 도표이다. 많은 양의 데이터 포인트가 그래프에 표시되면 점들이 서로 겹치게 되어 데이터의 분포나 특정 패턴을 파악하기 어려울 수 있다.
    - 이럴 때는 hexbin plot, 2D 히스토그램, KDE plot과 같은 다른 시각화 방법을 고려해한다.

```python
# 팁 금액과 총 청구금액의 연관성 시각화, 단 흡연 여부에 따라 구분
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="smoker")
plt.title("Tip Amount And Total Bill by Smoking")
plt.show()
```

## Line Plot

### 주요 특징

- **간결성**
    - 선 그래프는 간결하면서도 정보의 복잡성을 잘 전달할 수 있다.
- **시계열 데이터 표현**
    - 시간의 흐름에 따른 데이터의 변화를 직관적으로 표현한다.
    - 매출에 대한 월별 변화
- **변화와 추세에 대한 시각화**
    - 연속적인 데이터의 상승, 하락, 안정 등의 추세를 명확하게 보여준다.
    - 주가 변화
- **다중 데이터 비교**
    - 하나의 기준에 대해 여러개의 데이터를 한 그래프에 표시할 수 있다.
    - 두 회사의 매출 비교
- **적합한 데이터**
    - 시계열 데이터

---
### **한계**

- **대량의 데이터**
    - 산점도와 마찬가지로 매우 많은 데이터는 선 그래프가 지나치게 복잡해져서 정보의 해석이 어려울 수 있다.
- **데이터의 연속성 가정**
    - 선 그래프는 데이터 포인트 사이에 연속성이 있다고 가정한다.
    - 연속적이지 않은 범주형 데이터를 선으로 연결했을 때 데이터 해석이 달라질 수 있다.
- **누락된 데이터의 처리**
    - 데이터가 누락되거나 간격이 있을 때 선 그래프가 어떻게 그려지는지에 대한 문제가 발생할 수 있습니다.
- **간격이 불규칙한 데이터의 처리**
    - 데이터 간 간격이 일정하지 않을 경우 선 연결이 왜곡될 수 있다.

```python
# 예제 데이터 생성
flight_df = sns.load_dataset("flights")

# 기본 라인 플롯
sns.lineplot(flight_df, x="year", y="passengers")
plt.title("Basic Line Plot")
plt.show()
```

```python
# 월별 승객 수
sns.lineplot(data=flight_df, x="month", y="passengers")

plt.show()
```

## FacetGrid

### 특징

- 데이터 분석 시 특정 변수에 대해 다양한 측면에서 한 눈에 비교할 때 유용한 시각화 도구
- 데이터를 여러 서브플롯으로 나누어서 시각화할 수 있다.

```python
tips_df = sns.load_dataset('tips')
```

```python
grid = sns.FacetGrid(
    data = tips_df,
    col = 'time',   # 구분을 위한 범주형 데이터
)

# 시각화
grid.map(plt.hist, 'total_bill')

plt.show()
```

### (1) 타이타닉 데이터 시각화

1. 타이타닉 데이터프레임 불러오기    
2. `age` 와 `fare` 의 관계를 Scatter Plot으로 시각화한다.
3. `age` 와 `fare` 의 관계를 `survived` 로 구분해서 Scatter Plot으로 시각화한다.

```python
titanic = sns.load_dataset("titanic")
print(titanic.head())
```

```python
# age 와 fare 의 관계를 Scatter Plot으로 시각화한다.
sns.scatterplot(data = titanic, x = 'age', y = 'fare')
plt.show()
```

```python
# age 와 fare 의 관계를 survived 로 구분해서 Scatter Plot으로 시각화한다.
sns.scatterplot(data = titanic, x = 'age', y = 'fare', hue = 'survived')
plt.show()
```

### (2) 다이아몬드 데이터 시각화

1. 데이터프레임 생성
2. `price` 와 `carat` 의 관계를 Scatter Plot으로 시각화한다.
3. `price` 와 `carat` 의 관계를 `color` 로 구분해서 Scatter Plot으로 시각화한다.
4. `price` 와 `carat` 의  관계를 `cut` 로 구분해서 Scatter Plot으로 시각화한다.
5. `price` 와 `carat` 의 관계를 `cut` 품질에 따라 구분해서 다중 시각화한다.

```python
diamonds = sns.load_dataset("diamonds")
print(diamonds.head())
```

```python
# price 와 carat 의 관계를 Scatter Plot으로 시각화한다.
sns.scatterplot(data = diamonds, x = 'price', y = 'carat')
plt.show()
```

```python
# price 와 carat 의 관계를 color 로 구분해서 Scatter Plot으로 시각화한다.
sns.scatterplot(data = diamonds, x = 'price', y = 'carat', hue = 'color')
plt.show()
```

```python
# price 와 carat 의 관계를 cut 로 구분해서 Scatter Plot으로 시각화한다.
sns.scatterplot(data = diamonds, x = 'price', y = 'carat', hue = 'cut')
plt.show()
```

```python
# price 와 carat 의 관계를 cut 품질에 따라 구분해서 다중 시각화한다.
grid = sns.FacetGrid(data = diamonds, col = 'cut')
grid.map(sns.scatterplot, 'price', 'carat')
plt.show()
```
